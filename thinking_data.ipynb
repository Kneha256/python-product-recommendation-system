import json

## load data in read mode 

```python
def get_data(file):
    with open(file, "r") as f:
        data=json.load(f)
    return data
```

```python
data=get_data("store_data.json")
```

## Task1: clearing and structuring of data

```python
def clean_data(data):
    #convert rating data into int
    str_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        #cleaned rating - data consistancy
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in str_to_num:
            raw_rating = str_to_num[raw_rating]
        user["rating"] = raw_rating

        #handle missing data
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = None
            
        #deduplicated
        if user["name"].strip() in unique_users:
            continue
        else:
            unique_users.add(user["name"])
            cleaned_data.append(user)

    return cleaned_data
    
```

```python
data = clean_data(data)
#cleaned_data = clean_data(data)
#print(cleaned_data)
```

## task2: get meaningful insights

```python
def get_insight(data):

    #avg rating
    total_rating=0
    for user in data:
        total_rating += float(user["rating"])
    print(f"Average rating is : {total_rating/len(data)}")

    # % of poor 
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"% of users with poor ratings is : {poor_rating/len(data) * 100}%")        
```

```python
get_insight(data)
```

## Task3: Add code based Recommendation Features

```python
def get_recommendation(data):
    recommendation = []

    for user in data:
        curr_recc = {}
        curr_recc["name"] = user["name"]
        if (float(user["rating"]) >= 4):
            curr_recc["brand"] = "Apple"
        else:
            curr_recc["brand"] = "Samsung"

        recommendation.append(curr_recc)
    return recommendation
```

```python
get_recommendation(data)
```
